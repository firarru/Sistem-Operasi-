```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Cek file yang tersedia dalam dataset
files = os.listdir(path)
print("Files in dataset:", files)

dataset_path = os.path.join(path, "creditcard.csv")  # Sesuaikan dengan file yang ada
df = pd.read_csv(dataset_path)

# Tampilkan 5 baris pertama
df.head(10)
```

```python
import matplotlib.pyplot as plt
 Assuming 'df' is your DataFrame as defined in the previous code

df['Class'].value_counts().plot(kind='bar')
plt.title('Distribution of Class')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame as defined in the previous code

# Separate dataframes for each class
class_0 = df[df['Class'] == 0]
class_1 = df[df['Class'] == 1]

# Plotting histograms for each class
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(class_0['Amount'], bins=50)  # Adjust 'bins' as needed
plt.title('Histogram of Amount (Class 0)')
plt.xlabel('Amount')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)
plt.hist(class_1['Amount'], bins=50) # Adjust 'bins' as needed
plt.title('Histogram of Amount (Class 1)')
plt.xlabel('Amount')
plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```
